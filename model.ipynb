```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('keypoint.csv', header=None, names=["Class"]+[i for i in range(66)])
df.head()
```

```python
df=df.sample(frac=1)
df
```

```python
X=df.drop(['Class'], axis=1)
X
```

```python
X.shape
```

```python
y=df["Class"]
y
```

```python
X=np.array(X, dtype='float32')
```

```python
from sklearn.preprocessing import LabelEncoder

y=LabelEncoder().fit_transform(y)
y
```

```python
y=np.array(y).reshape(-1,1)
```

```python
from keras import models, layers

model=models.Sequential()
model.add(layers.Dense(512, activation="relu"))
model.add(layers.Dense(128, activation="LeakyReLU"))
model.add(layers.Dense(6, activation="softmax"))
```

```python
model.compile(
    optimizer="Adam", 
    loss="sparse_categorical_crossentropy",
    metrics=['accuracy'],
)
```

```python
model.fit(X, y, epochs=100)
```

```python
preds=model.predict(X)
preds=np.argmax(preds, axis=1)
preds
```

```python
y
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score

sns.heatmap(confusion_matrix(y, preds), annot=True)
```

Save the model

```python
MODEL_PATH='./MODEL/yoga/model.hdf5'
```

```python
model.save(MODEL_PATH, include_optimizer=False)
```

```python
import tensorflow as tf
```

```python
tflite_save_path='./MODEL/yoga/model.tflite'
```

```python
converter=tf.lite.TFLiteConverter.from_keras_model(model)
converter.optimizations=[tf.lite.Optimize.DEFAULT]
```

```python
tflite_quantized_model=converter.convert()
```

```python
open(tflite_save_path, 'wb').write(tflite_quantized_model)
```

Testing the TFLite Model

```python
interpreter = tf.lite.Interpreter(model_path=tflite_save_path)
interpreter.allocate_tensors()
```

```python
input_details = interpreter.get_input_details()
output_details = interpreter.get_output_details()
```

```python
interpreter.set_tensor(input_details[0]['index'], np.array([X[2]]))
```

```python
interpreter.invoke()
tflite_results=interpreter.get_tensor(output_details[0]['index'])
```

```python
np.squeeze(tflite_results)
```
